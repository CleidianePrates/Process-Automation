# Automação Web e Busca de Informações com Python

#### Desafio: 

Trabalhamos em uma importadora e o preço dos nossos produtos é vinculado a cotação de:
- Dólar
- Euro
- Ouro

Precisamos pegar na internet, de forma automática, a cotação desses 3 itens e saber quanto devemos cobrar pelos nossos produtos, considerando uma margem de contribuição que temos na nossa base de dados.

Base de Dados: https://drive.google.com/drive/folders/1KmAdo593nD8J9QBaZxPOG1yxHZua4Rtv?usp=sharing

Para isso, vamos criar uma automação web:

- Usaremos o selenium
- Importante: baixar o webdriver

```python
# -- ! pip install selenium
```

```python
#chrome -> chromedriver

from selenium import webdriver #controlar o navegador
from selenium.webdriver.common.keys import Keys #usar teclado
from selenium.webdriver.common.by import By #Localiar os itens no navegador

#Criar o Navegador
navegador = webdriver.Chrome()


#Entraria no google e pesquisaria a cotação do dólar
navegador.get("https://www.google.com/")

#pegar a cotação do dólar
#--navegador.find_element(qual o paramento quer usar, qual o código do parametro)
navegador.find_element(By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input").send_keys("cotação dolar")

#--no selenium, escrever chama send_Keys

navegador.find_element(By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input").send_keys(Keys.ENTER)
#--no selenium, enter = Keys.ENTER

cotacao_dolar = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
#--no selenium, get_attribute pega o atributo no XPATH do site

print(cotacao_dolar)
#Entraria no google e pesquisaria a cotação do euro
navegador.get("https://www.google.com/")

navegador.find_element(By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input").send_keys("cotação euro")

#--buscou euro no google
navegador.find_element(By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input").send_keys(Keys.ENTER)

#pegar a cotação do euro

cotacao_euro = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

print(cotacao_euro)

#Cotação ouro
#--entrar no https:/www.melhorcambio.com/ouro-hoje
navegador.get("https:/www.melhorcambio.com/ouro-hoje")


#pegaria a cotação do ouro

cotacao_ouro = navegador.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute('value')

cotacao_ouro = cotacao_ouro.replace(",",".")

print(cotacao_ouro)

navegador.quit()
#importar a base de dados




```

### Agora vamos atualiza a nossa base de preços com as novas cotações

- Importando a base de dados

```python
import pandas as pd

tabela = pd.read_excel("produtos.xlsx")
display(tabela)
```

- Atualizando os preços e o cálculo do Preço Final

```python
#atualizar as cotações na base de dados
tabela["Moeda"] == "Dólar"
tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(cotacao_dolar)
tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacao_euro)

tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacao_ouro)

#-- '.loc' localiza as linhas e colunas = tabela.loc[linhas, clunas] = type(informação a ser inserida)

#--preço de compra = preço original * cotação
tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"]
#--preço de venda = cotação * margem
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"]


display(tabela)
```

### Agora vamos exportar a nova base de preços atualizada

```python
#exportar esa base de dados para ter o resultado atualizado
tabela.to_excel('ProdutosNovo.xlsx',index=False)
```
